```python
import matplotlib.pyplot as plt
import numpy as np

def main():
  x = np.linspace(0,2*np.pi, 200)
  f = input("Enter('sin', 'cos', or 'tan'): ")

  if f == 'sin':
    y = np.sin(x)
    plt.figure(figsize = (20, 10))
    plt.plot(x,y)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()

  elif f == 'tan':
    x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)
    y = np.tan(x)
    plt.figure(figsize = (20, 10))
    plt.plot(x, y)
    plt.ylim(-5, 5)


  elif f == 'cos':
    y = np.cos(x)
    plt.figure(figsize = (20, 10))
    plt.plot(x,y)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()

  else:
    print("Invalid, enter again. ")



main()
```
