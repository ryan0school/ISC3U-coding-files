```python
import cmath
import matplotlib.pyplot as plt
import numpy as np

try:
  e = input("Enter polynomial equation using 'x' as the variable ': ")
  x = np.linspace(-10, 10, 400)
  y = eval(e)

  #plotting graph
  plt.plot(x, y)
  plt.xlabel('x')
  plt.ylabel('y')
  plt.title(f'Polynomial Equation: {e}')
  plt.grid(True)
  plt.show()
  plt.figure(figsize=(10,6))

except (SyntaxError, NameError, TypeError):
  print("Invalid, input again")
```
