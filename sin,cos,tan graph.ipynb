```python
import matplotlib.pyplot as plt
import numpy as np

def main():
  x = np.linspace(0,2*np.pi, 200)
  f = input("Enter('sin', 'cos', or 'tan'): ")

  if f == 'sin':
    y = np.sin(x)

  elif f == 'tan':
    y = np.tan(x)

  elif f == 'cos':
    y = np.cos(x)

  else:
    print("Invalid, enter again. ")

  plt.figure(figsize = (20, 10))
  plt.plot(x,y)
  plt.xlabel('x')
  plt.ylabel('y')
  plt.show()

main()
```
